```python
#todos los puntos
sns.lmplot(data=df, x='x', y='y')
```

```python
#subplots con parametro de columna
sns.lmplot(data=df, x='x', y='y', col='dataset')
```

```python
sns.lmplot(data=df, x='x', y='y', col='dataset', hue='dataset', col_wrap=2)
```

```python
x = df[df.dataset == 'I'].x
y = df[df.dataset == 'I'].y
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y)
```

```python
#coeficiente de correlaçao
r, p_value = stats.pearsonr(x,y)
r
```

```python
p_value
```

```python
#66% de la variacion de I esta explicada por la variacion de x
R2 = r ** 2
R2
```

```python
sx = x.std()
sy = y.std()
```

```python
#intercept, medias
mx = x.mean()
my = y.mean()
```

```python
b = r * sy / sx
b
```

```python
#equacion de una recta
y = b * x + a
```

```python
my = b * mx + a
```

```python
a = my - b * mx
```

```python
# para x = 4 y b = 14 qual o valor de I
plt.plot((4, 14), (b*4+a,b*14+a))
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y)
plt.plot((4, 14), (b*4+a,b*14+a), c='r')
```

```python
stats.pearsonr (x,y)
```

```python
stats.spearmanr (x,y)
```

```python
stats.kendalltau(x,y)
```

```python
df2 = sns.load_dataset ('penguins')
```

```python
df2.dropna(inplace=True)
```

```python
df2.head()
```

```python
#cual es la correlacion entre todas la variables numericas
sns.pairplot(data=df2)
```

```python
sns.pairplot(data=df2, hue='sex')
```

```python
#antes de calcular tenemos que tener claro entre que queremos calcular los coeficientes de correlacion
```

```python
#correlacion negativa
sns.lmplot(data=df2, x='body_mass_g', y='bill_depth_mm', hue='sex')
```

```python
#correlacion positiva, dentro de la misma especie
sns.lmplot(data=df2, x='body_mass_g', y='bill_depth_mm', hue='species')
```

```python
# de una especie especifica
sns.pairplot(data=df2[df2.species=='Gentoo'], hue='sex')
```

```python
df2[df2.species=='Gentoo'].corr()
```

```python
df2[df2.species=='Gentoo'].corr(method='spearman')
```

```python
#mapa de color
sns.heatmap(df2[df2.species=='Gentoo'].corr(), annot=True, cmap='Blues')
```

```python
#scaterplot 
sns.jointplot(data=df2, x='body_mass_g', y='bill_depth_mm', hue='species')
```
