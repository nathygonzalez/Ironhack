```python
lluvia = respuesta ['pronostico']['manana']['prob_precipitacion']
```

```python
lluvia = list(map (int, lluvia))
```

```python
for x in lluvia:
    if x > 50:
        print ("Mañana va a llover.")
        break
```

```python
# in boolean
map(lambda x: x > 50, lluvia)
```

```python
any(list(map(lambda x: x > 50, lluvia)))
```

```python
sum = any(list(map(lambda x: x > 50, lluvia)))
if sum > 0:
    print ("Mañana va a llover.")
else:
    print ("Mañana barbacoa")
```
