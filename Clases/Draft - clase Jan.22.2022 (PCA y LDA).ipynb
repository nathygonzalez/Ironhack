```python
from sklearn.linear_model import LogisticRegression
```

```python
# hay que aumentar el num. de iteraciones
model = LogisticRegression(max_iter=10000)
```

```python
X = cancer.data
y = cancer.target
model.fit(X, y)
```

```python
# cuan buenos es el modelo, cual la correlacion con los datos - es muy bueno, cerca de 1
model.score(X,y)
```

# PCA

```python
from sklearn.decomposition import PCA, NMF
```

```python
# Hay que estandarizar los datos, PCA es la rotacion
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Estandarizar = restar la media y dividir la (?deviación) típica
X = StandardScaler().fit_transform(cancer.data)
```

```python
pca = PCA()
```

```python
# lo entrenaremos con nuestros datos, entrenamiento con los componentes principales
pca.fit(X)
```

```python
dir(pca)
```

```python
# con 3 variables tengo 80
plt.plot(pca.explained_variance_ratio_)
```

```python
# esta es la suma acumulada, 3 variables explico 80% de la varianza de mis datos
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
pca.components_
```

```python
pd.DataFrame(pca.components_, columns=cancer.feature_names)
```

```python
# de la 1a componente que es lo que este recogendo / a que variables esta haciendo referencia
sns.heatmap(pd.DataFrame(pca.components_, columns=cancer.feature_names), center=0, cmap='coolwarm')
```

```python
loadings = pca.transform(X)
```

```python
X2 = loadings[:,0:3]
```

```python
# esta es otra matriz, representacion de mis datos en el nuevo espacio (loadings)
# aprendo el nuevo espacio y lo proyecto en el nuevo espacio
X2.shape
```

```python
# un resultado muy similar al alterior pero solo con 3 componentes
model.fit(X2,y)
model.score(X2,y)
```

```python
# crear un array de colores en funcion del target
np.where(cancer.target, 'b')
```

```python
# con la 1a y 2a columna de loadings - la separacion es un poco más definida que la anterior, pero reduciendo de 30 a 3 variables
plt.scatter(loadings[:,0], loadings[:,1], c=cancer.target)
```

# LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# instanciamos nuestro metodo
lda = LinearDiscriminantAnalysis()
```

```python
dir (lda)
```

```python
x = lda.fit_transform(X,y)
```

```python
# me devuelve solo 1 componente, la que mas me separa mis datos
x.shape
```

```python
# el resultado se ve mejor todavia que los anteriores
lda.score(X,y)
```

```python
# puedo utilizarlo como un predictor mas, con logistic regression para predecir el resultado
y_pred = lda.predict(X)
```

```python
from sklearn.metrics import classification_report
```

```python
# resultado muy bueno
print(classification_report(y, y_pred))
```

```python
# para cada una de mis observaciones - con la longitud de mis datos , de mi muestra
# colorear por mi variable objetivo
plt.scatter (range(len(df)), x, c=y)
```
