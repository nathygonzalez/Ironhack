```python
%%sql
with firstorderpercustomer as
(select IdCliente, min(FechaPedido) as firstorderdoneon
from Pedidos
group by IdCliente)
select IdCliente,
firstorderdoneon,
substr(firstorderdoneon, 0,8) as FirstOrderMonth,
julianday (firstorderdoneon) days
from Pedidos orders
left join firstorderpercustomer
on firstorderpercustomer.IdCliente =
orders.IdCliente
```

```python
%%sql
with ﬁrstorderpercustomer as
(select IdCliente, min(FechaPedido) as ﬁrstorderdoneon
from Pedidos
group by IdCliente)
select
   ﬁrstorderdoneon,
   substr(ﬁrstorderdoneon, 0,8) as FirstOrderMonth,
   julianday(ﬁrstorderdoneon) julianFirstOrderDate,
   FechaPedido,
   julianday(FechaPedido) JulianOrderDate,
(julianday(FechaPedido) - julianday(ﬁrstorderdoneon)) CohortDay,
round((julianday(FechaPedido) - julianday(ﬁrstorderdoneon))/30) as cohortmonth
from Pedidos orders
left join ﬁrstorderpercustomer
on ﬁrstorderpercustomer.IdCliente =
orders.IdCliente
```

```python
%%sql
with ﬁrstorderpercustomer as
(select IdCliente, min(FechaPedido) as ﬁrstorderdoneon
from Pedidos
group by IdCliente)
select
   substr(ﬁrstorderdoneon, 0,8) as FirstOrderMonth,
   round((julianday(FechaPedido) - julianday(ﬁrstorderdoneon))/30) as cohortmonth,
   count(distinct orders.IdCliente) num
from Pedidos orders
left join ﬁrstorderpercustomer
on ﬁrstorderpercustomer.IdCliente =
orders.IdCliente
group by FirstOrderMonth, cohortmonth
```
