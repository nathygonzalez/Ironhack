```python
import numpy as pd
import numpy as np
```

```python
n = 1000
```

```python
correcto = np.random.choice ([1,2,3], n)
correcto
```

```python
seleccion = np.random.choice ([1,2,3], n)
```

```python
[c==s for c,s in zip (correcto, seleccion)]
```

```python
sum([c==s for c,s in zip (correcto, seleccion)]) / n
```

```python
# el resultado es un conjunto, hay que sacar uno 
abierta = [({1,2,3} - {c,s}).pop() for c,s in zip(correcto, seleccion)]
```

```python
cambio = [({1,2,3} - {c,s}).pop() for c,s in zip(abierta, seleccion)]
```

```python
sum([c==s for c,s in zip (correcto, cambio)]) / n
```
